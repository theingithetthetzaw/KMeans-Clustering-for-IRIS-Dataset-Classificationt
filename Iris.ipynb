```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
le.fit(Y)
Y=le.transform(Y)
Y=np.array(Y)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=12)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB #import model
gnb=GaussianNB()
gnb.fit(X_train,Y_train) 
```

```python
y_pred_train= gnb.predict(X_train)
print("Number of mislabeled points out of a total training set : %d" %(Y_train !=y_pred_train).sum())
print(y_pred_train)
print(Y_train)
```

```python
y_pred_test= gnb.predict(X_test)
print("Number of mislabeled points out of a total training set : %d" %(Y_train !=y_pred_train).sum())
print("Number ....: %d" %(Y_test != y_pred_test).sum())
```

```python
test=[6.4,3.2,4.5,1.5]
test=np.array(test)
tt=test.reshape(1,-1)

t=gnb.predict(tt)
print(t)
```
